```python
from sklearn.datasets import fetch_openml

feature, target = fetch_openml("mnist_784",return_X_y=True)
```

```python
target = target.astype(int)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt


plt.figure(figsize=(10,5))
for i in range(10):
    subplot = plt.subplot(2,5, i+1)
    subplot.imshow(feature[target==i].values[0].reshape(28,28))
plt.show()
```

```python
from sklearn.model_selection import train_test_split
import numpy as np

train_feature, test_feature, train_target, test_target = train_test_split(feature,target,random_state=42, stratify=target)
train_target = np.array(train_target)
test_target = np.array(test_target)
```

```python
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(n_jobs=-1, random_state=42)
clf.fit(train_feature,train_target)
```

```python
y_pred = clf.predict(test_feature)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

print("Accuracy :",accuracy_score(test_target, y_pred))
print(confusion_matrix(test_target, y_pred,labels=[9,8,7,6,5,4,3,2,1,0]))
```

```python
from sklearn.cluster import KMeans

kmean = KMeans(n_clusters = 10, random_state=42)
kmean.fit(train_feature)
```

```python
y_pred = kmean.predict(test_feature)
```

```python
results = {}
for i in range(len(y_pred)):
    if y_pred[i] not in results:
        results[y_pred[i]] = {}
        results[y_pred[i]]['cluster']=1
        results[y_pred[i]]['real']=[]
    else:
        results[y_pred[i]]['cluster']+=1
    results[y_pred[i]]['real'].append(test_target[i])
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

for k in range(2,50):
    kmean = KMeans(n_clusters = k, random_state=42)
    kmean.fit(train_feature)
    print(k, silhouette_score(train_feature, kmean.labels_))
```

```python
from sklearn.cluster import KMeans
kmean = KMeans(n_clusters = 42, random_state=42)
```

```python
train_feature_cluster = kmean.fit_predict(train_feature)
test_feature_cluster = kmean.predict(test_feature)
```

```python
import numpy as np

train_feature_cluster = np.array(train_feature_cluster).reshape(-1, 1)
test_feature_cluster = np.array(test_feature_cluster).reshape(-1,1)
```

```python
from sklearn.ensemble import RandomForestClassifier

clf2 = RandomForestClassifier(n_jobs=-1, random_state=42)
clf2.fit(train_feature_cluster,train_target)
```

```python
y_pred = clf2.predict(test_feature_cluster)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

print("Accuracy :",accuracy_score(test_target, y_pred))
print(confusion_matrix(test_target, y_pred,labels=[9,8,7,6,5,4,3,2,1,0]))
```
