```python
import pandas as pd

df = pd.read_csv("depressive_disorder_dataset.csv")
```

```python
df.head()
```

```python
df.loc[df['우울증 현재 유병 여부']=='의사진단 받지 않음','우울증 현재 유병 여부'] = None
```

```python
df.dropna(inplace=True)
df.reset_index(inplace=True, drop=True)
```

```python
df
```

```python
df.loc[df['성별']=='남성','성별']=1
df.loc[df['성별']=='여성','성별']=0
df['성별'] = df['성별'].astype('int')
```

```python
df.loc[df['우울증 현재 유병 여부']=='있음','우울증 현재 유병 여부']=1
df.loc[df['우울증 현재 유병 여부']=='없음','우울증 현재 유병 여부']=0
df['우울증 현재 유병 여부'] = df['우울증 현재 유병 여부'].astype('int')
```

```python
df = pd.get_dummies(df)
```

```python
df
```

```python
features = df.drop(['우울증 현재 유병 여부'], axis=1)
```

```python
target = df['우울증 현재 유병 여부']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(features, target, random_state=42, stratify=target)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.svm import SVC

clf = SVC()
clf.fit(X_train_scaled, y_train)
```

```python
pred = clf.predict(X_test_scaled)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

print("Accuracy :",accuracy_score(y_test, pred))
print("Precision :",precision_score(y_test, pred))
print("Recall :",recall_score(y_test, pred))
print("F1 score :",f1_score(y_test, pred))
```

```python
from sklearn.metrics import confusion_matrix

print(confusion_matrix(y_test, pred))
```

```python
from sklearn.metrics import confusion_matrix

print(confusion_matrix(y_test, pred,labels=[1,0]))
```

```python
from sklearn.svm import LinearSVC

clf = LinearSVC()
clf.fit(X_train_scaled, y_train)
```

```python
pred = clf.predict(X_test_scaled)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

print("Accuracy :",accuracy_score(y_test, pred))
print("Precision :",precision_score(y_test, pred))
print("Recall :",recall_score(y_test, pred))
print("F1 score :",f1_score(y_test, pred))
```

```python
from sklearn.metrics import confusion_matrix

print(confusion_matrix(y_test, pred,labels=[1,0]))
```
