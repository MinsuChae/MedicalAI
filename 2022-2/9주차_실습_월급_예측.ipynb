```python
import pandas as pd

df = pd.read_csv("depressive_disorder_dataset.csv")
```

```python
df.head()
```

```python
df.loc[df['가구원 수']!=1,'가구원 수'] = None
df.loc[df['경제활동 상태']!='취업자','경제활동 상태'] = None
```

```python
df.dropna(inplace=True)
df.reset_index(inplace=True, drop=True)
```

```python
df
```

```python
df.loc[df['성별']=='남성','성별']=1
df.loc[df['성별']=='여성','성별']=0
df['성별'] = df['성별'].astype('int')
```

```python
features = df[['성별','나이','교육수준','표준직업분류 대분류 코드','종사자지위','임금근로자 상세','정규직 여부']]
target = df['월 가구총소득']
```

```python
features
```

```python
features = pd.get_dummies(features)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(features, target, random_state=42)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.svm import SVR

clf = SVR()
clf.fit(X_train_scaled, y_train)
```

```python
pred = clf.predict(X_test_scaled)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

print("MAE :",mean_absolute_error(y_test,pred))
print("RMSE :",np.sqrt(mean_squared_error(y_test,pred)))
```

```python
from sklearn.svm import SVR

clf = SVR(kernel='linear')
clf.fit(X_train_scaled, y_train)
```

```python
pred = clf.predict(X_test_scaled)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

print("MAE :",mean_absolute_error(y_test,pred))
print("RMSE :",np.sqrt(mean_squared_error(y_test,pred)))
```

```python
from sklearn.svm import LinearSVR

clf = LinearSVR()
clf.fit(X_train_scaled, y_train)
```

```python
pred = clf.predict(X_test_scaled)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

print("MAE :",mean_absolute_error(y_test,pred))
print("RMSE :",np.sqrt(mean_squared_error(y_test,pred)))
```
